```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
#  url 
travel_videos_url = "https://www.youtube.com/results?search_query=travel"

# Setting up driver to fetch content from youtube
driver = webdriver.Chrome()
driver.get(travel_videos_url)
```

```python
# Locating elements using XPath
user_data = driver.find_elements(By.XPATH, '//*[@id="video-title"]')

# Empty list to store linkf
links = []
for element in user_data:
    link = element.get_attribute('href')
    if link: 
        links.append(link)

print(len(links))
```

```python
df = pd.DataFrame(columns=['link','title','description','category'])
```

```python
wait = WebDriverWait(driver, 10)

# Storing the video category name (travel)
v_category = "travel"

# looping through the youtube videos links
for x in links:

    # Directing the driver to open each link
    driver.get(x)

    # Extracting the video Id
    v_id = x.strip('https://www.youtube.com/watch?v=')

    # Extracting the video title
    v_title = wait.until(EC.presence_of_element_located(
                   (By.CSS_SELECTOR, "h1.title yt-formatted-string"))).text
    
    # Extracting the video description
    v_description = wait.until(EC.presence_of_element_located(
        (By.CSS_SELECTOR, "div#description yt-formatted-string"))).text
    
    # Adding the data to a dataframe
    df.loc[len(df)] = [v_id, v_title, v_description, v_category]
```
